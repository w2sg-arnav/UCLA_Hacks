```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = r"C:\Users\aaria\Downloads\merged_train_data.csv"
sample_size = 1000000

try:
    advertiser_data = pd.read_csv(file_path, nrows=sample_size)
except FileNotFoundError:
    print(f"Error: File '{file_path}' not found.")
    exit()
except pd.errors.EmptyDataError:
    print(f"Error: File '{file_path}' is empty or cannot be read as CSV.")
    exit()
except Exception as e:
    print(f"Error: {e}")
    exit()

all_users_age = advertiser_data['age']
clicked_data = advertiser_data[advertiser_data['label'] == 1]
clicked_users_age = clicked_data['age']

plt.figure(figsize=(10, 6))
sns.kdeplot(all_users_age, label='All Users', shade=True)
sns.kdeplot(clicked_users_age, label='Users Who Clicked Ads', shade=True)

plt.title('Age Group Distribution of Users Who Clicked Ads vs. All Users')
plt.xlabel('Age')
plt.ylabel('Density')

plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = r"C:\Users\aaria\Downloads\merged_train_data.csv"

try:
    advertiser_data = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"Error: File '{file_path}' not found.")
    exit()
except pd.errors.EmptyDataError:
    print(f"Error: File '{file_path}' is empty or cannot be read as CSV.")
    exit()
except Exception as e:
    print(f"Error: {e}")
    exit()

clicked_data = advertiser_data[advertiser_data['label'] == 1]

features = ['city_rank', 'residence', 'city']

fig, axes = plt.subplots(len(features), 1, figsize=(10, 8 * len(features)))

for i, column in enumerate(features):
    ax = axes[i]
    sns.kdeplot(advertiser_data[column], label='All Users', shade=True, ax=ax)
    sns.kdeplot(clicked_data[column], label='Users Who Clicked Ads', shade=True, ax=ax)
    ax.set_title(f'{column.capitalize()} Distribution')
    ax.set_xlabel(column)
    ax.set_ylabel('Density')
    ax.legend()

plt.tight_layout()
plt.show()
```
